```python
import tkinter as tk
from PIL import Image, ImageTk
import cv2
import numpy as np
import joblib
import pyscreenshot as ImageGrab

window = tk.Tk()
window.title("Handwritten Digit Classification")
window.geometry("800x600")
window.resizable(0, 0)

# Open the image file
load1 = Image.open("HANDWRITTENDIGITCLASSIFICATION.png")

# Create a canvas
canvas1 = tk.Canvas(window, width=500, height=250, bg="ivory")
canvas1.place(x=5, y=120)

# Resize the image to fit the canvas dimensions
canvas_width, canvas_height = canvas1.winfo_width(), canvas1.winfo_height()

# Assuming you already opened the image using Image.open and resized it
canvas_image = ImageTk.PhotoImage(load1.resize((canvas_width, canvas_height)))

# Display the image as the canvas background using the created variable
canvas1.create_image(0, 0, image=canvas_image, anchor='nw')


l1 = tk.Label(canvas1,text="Digit",font=('Algerian',20))
l1.place(x=5,y=0)

t1 = tk.Entry(canvas1,width=20, border=5)
t1.place(x=150,y=5)

b1 = tk.Button(canvas1,text="1. Open paint and capture the screen",font=('Algerian',15),bg="orange",fg="black")
b1.place(x=5,y=50)

b2 = tk.Button(canvas1,text="2. Generate dataset",font=('Algerian',15),bg="pink",fg="blue")
b2.place(x=5,y=100)

b3 = tk.Button(canvas1,text="3. Train the model and calculate accuracy",font=('Algerian',15),bg="green",fg="white")
b3.place(x=5,y=150)

def prediction():
    # Load the model
    # model = joblib.load("MINOR PROJECT\model")

    img = ImageGrab.grab(bbox=(130,500,500,700))
    img.save("paint.png")
    im = cv2.imread("paint.png")
    load = Image.open("paint.png")
    load = load.resize((280,280))
    photo = ImageTk.PhotoImage(load)
    
    # Labels can be text or images
    img_label = tk.Label(canvas3, image=photo, width=280, height=280)
    img_label.image = photo
    img_label.place(x=0, y=0)
    
    im_gray = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
    im_gray = cv2.GaussianBlur(im_gray, (15,15), 0)
    
    # Threshold the image
    ret, im_th = cv2.threshold(im_gray, 100, 255, cv2.THRESH_BINARY)
    roi = cv2.resize(im_th, (28,28), interpolation=cv2.INTER_AREA)
    
    rows, cols = roi.shape
    
    x = []
    
    # Add pixel one by one into data array
    for i in range(rows):
        for j in range(cols):
            k = roi[i,j]
            if k > 100:
                k = 1
            else:
                k = 0
            x.append(k)
    x = np.array(x).reshape(1, -1)  # Reshape the input for prediction
    
    # Make prediction (uncomment the line below once model is loaded)
    # prediction = model.predict(x)
    
    # For now, let's create a placeholder prediction
    prediction = np.random.randint(0, 10)  # Assuming 0 to 9 are possible predictions
    
    # Printing the prediction in canvas3 window
    prediction_text = f"Prediction = {prediction}"
    a1 = tk.Label(canvas3, text="Prediction = ", font=("Algerian", 20))
    a1.place(x=5, y=350)
    
    b1 = tk.Label(canvas3, text=prediction_text, font=("Algerian", 20))
    b1.place(x=200, y=350)

b4 = tk.Button(canvas1, text="4. Live prediction", font=("Algerian", 15), bg="white", fg="red", command=prediction)
b4.place(x=5, y=200)

canvas2 = tk.Canvas(window, width=500, height=270, bg="black")
canvas2.place(x=5, y=380)


def activate_paint(e):
    global lastx, lasty
    canvas2.bind('<B1-Motion>',paint)
    lastx, lasty = e.x,e.y

def paint(e):
    global lastx,lasty
    x,y = e.x,e.y
    canvas2.create_line((lastx,lasty,x,y),width=30,fill="white")
    lastx,lasty = x,y
    
canvas2.bind('<1>',activate_paint)

def clear():
    canvas2.delete("all")

btn = tk.Button(canvas2,text="clear",fg="white",bg="red",command=clear)
btn.place(x=0,y=0)

canvas3 = Canvas(window,width=280,height=530,bg="green")
canvas3.place(x=515,y=120)



Frame1 = Frame(window,)
window.geometry("800x600")

# b4.config(command=predict_digit)  # Configure the button to call predict_digit function

window.mainloop()

```

```python
def train_save_accuracy():
    import pandas as pd
    from sklearn.utils import shuffle
    data = pd.read_csv('dataset.csv')
    data = shuffle(data)
    X = data.drop(["label"],axis=1)
    Y = data["label"]
    from sklearn.model_selection import train_test_split
    train_x,test_x,train_y,test_y = train_test_split(X,Y,test_size = 0.2)
    import joblib
    from sklearn.svm import SVC
    classifier = SVC(kernel="linear",random_state=6)
    classifier.fit(train_x,train_y)
    joblib.dump(classifier,"model/digit_recognizer")
    from sklearn import metrics
    prediction = classifier.predict(test_x)
#     print("Accuracy = ",metrics.accuracy_score(prediction,test_y))
```
